```python
import json
import os

from runtime import RunTime
```

```python
models = ['unet', 'resnet', 'densenet', 'hrnet']
pockets = [True, False]
cnt = 1

for model in models:
    for pocket in pockets:
                
        if pocket:
            base_model_name = 'nfbs_pocket_{}'.format(model)
            model_dir = '/tf/data/nfbs/models/pocketnet/{}_pocket'.format(model)
            prediction_dir = '/tf/data/nfbs/predictions/pocketnet/{}_pocket'.format(model)
            results_csv = '/tf/github/MIST/mist/results/nfbs_pocket_{}_results.csv'.format(model)
            user_params_json_file = os.path.join('/tf/github/MIST/mist/user_params', 
                                                 'nfbs_pocket_{}_user_params.json'.format(model))
            
        else:
            base_model_name = 'nfbs_full_{}'.format(model)
            model_dir = '/tf/data/nfbs/models/pocketnet/{}_full'.format(model)
            prediction_dir = '/tf/data/nfbs/predictions/pocketnet/{}_full'.format(model)
            results_csv = '/tf/github/MIST/mist/results/nfbs_full_{}_results.csv'.format(model)
            user_params_json_file = os.path.join('/tf/github/MIST/mist/user_params', 
                                                 'nfbs_full_{}_user_params.json'.format(model))
        
        user_params = {'raw_data_dir': '/tf/data/nfbs/raw',
                       'processed_data_dir': '/tf/data/nfbs/processed/tfrecord',
                       'base_model_name': base_model_name,
                       'model_dir': model_dir,
                       'prediction_dir': prediction_dir,
                       'raw_paths_csv': '/tf/github/MIST/mist/path_csv/nfbs_paths.csv',
                       'inferred_params': '/tf/github/MIST/mist/inferred_params/nfbs_inferred_params.json',
                       'results_csv': results_csv,
                       'modality': 'mr',
                       'mask': ['brainmask.nii.gz'], 
                       'images': {'t1': ['T1w.nii.gz']}, 
                       'labels': [0, 1],
                       'final_classes': {'BRAIN': [1]},
                       'loss': 'gdl', 
                       'model': model, 
                       'pocket': pocket, 
                       'gpu': 7, 
                       'folds': 3}

        
        with open(user_params_json_file, 'w') as outfile: 
            json.dump(user_params, outfile)
            
        train = RunTime(user_params_json_file)
        if cnt == 1:
            train.run(run_preprocess = False)
        else:
            train.run(run_preprocess = False)
            
        cnt += 1
```
